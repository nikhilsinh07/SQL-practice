```python
%%sql

select
orderdate,
date_part('year',orderdate),
date_part('month',orderdate),
date_part('day',orderdate)
from sales
limit 10
```

```python
%%sql
select
extract(year from orderdate) as extracted_year,
extract(month from orderdate) as extracted_month,
sum(s.quantity*s.netprice*s.exchangerate) as net_price
from sales s
group by orderdate
limit 10
```

```python
%%sql
select
CURRENT_DATE,
s.orderdate,
p.categoryname,
sum(s.quantity*s.netprice*s.exchangerate) as net_revenue
from sales s
left join product p on p.productkey=s.productkey
where (extract(year from orderdate)) > (extract(year from CURRENT_DATE) - 5)
group by p.categoryname,orderdate
order by s.orderdate,p.categoryname
```

```python
%%sql
select
CURRENT_DATE,
orderdate,
p.categoryname,
sum(s.quantity*s.netprice*s.exchangerate) as net_revenue
from sales s
left join product p on p.productkey=s.productkey
where
orderdate >= CURRENT_DATE-INTERVAL'5years'
group by s.orderdate,p.categoryname
order by s.orderdate,p.categoryname
```

```python
%%sql
select
extract( month from orderdate) as month,
sum(distinct s.quantity*s.netprice*s.exchangerate) as net_revenue_monthly
from sales s
group by month
```

```python
%%sql
select
date_part('year',orderdate) as order_year,
round(avg(extract(days from age( deliverydate,orderdate))),2) as processing_time,
sum(distinct s.quantity*s.netprice*s.exchangerate) as net_revenue_monthly
from sales s
where orderdate >= current_date - interval '5 years'
group by order_year
order by order_year
limit 10
```
